### **Data Wrangling Tools**

The data you choose may require reformatting, transformation or cleaning prior to visualization. You can use Pandas for data preparation. 

```
# READ FROM CSV INTO PANDAS DF
df = pd.read_csv('shot_logs.csv')

# Isolate variables of interest
df = df[['SHOT_CLOCK', 'DRIBBLES', 'TOUCH_TIME', 'SHOT_DIST', 'SHOT_RESULT', 'CLOSE_DEF_DIST', 'player_name']]

# Drop rows with missing values
df = df.dropna()
print(df)
```

```
# GRAPH 1: Distribution of shots over the shot clock period
plt.style.use('seaborn')
plt.hist(df['SHOT_CLOCK'], bins=8, color='#2185AC', rwidth=0.8)
plt.xlabel('Shot Clock (s)')
plt.ylabel('# of Shots')
plt.title('Distribution of Shots Over a 24 Second Period')
plt.show()
```

```
# GRAPH #2: Stacked Histogram of shots made vs missed over the shot clock period
df1 = df[df['SHOT_RESULT'] == 'made']
df2 = df[df['SHOT_RESULT'] == 'missed']
counts, bins, bars = plt.hist([df1['SHOT_CLOCK'], df2['SHOT_CLOCK']], bins=8, rwidth=0.8, stacked=True, color = ['b','r'], label=['made', 'missed'])
plt.xlabel('Shot Clock (s)')
plt.ylabel('# of Shots')
plt.title('Distribution of Shots Over a 24 Second Period')
plt.legend()
```

```
# GRAPH #3: STEPHEN CURRY - Stacked Histogram of shots made vs missed over the shot clock period
df_SC = df[df['player_name'] == 'stephen curry']
print(df_SC)
df1_SC = df_SC[df_SC['SHOT_RESULT'] == 'made']
df2_SC = df_SC[df_SC['SHOT_RESULT'] == 'missed']
counts_SC, bins_SC, bars_SC = plt.hist([df1_SC['SHOT_CLOCK'], df2_SC['SHOT_CLOCK']], bins=8, rwidth=0.8, stacked=True, color = ['b','r'], label=['made', 'missed'])
plt.xlabel('Shot Clock (s)')
plt.ylabel('# of Shots')
plt.title('Distribution of Stephen Curry\'s Shots Over a 24 Second Period')
plt.legend()
```

```
# GRAPH #4: Line graph of shot percentages over the shot clock period
x = []
y = []
for i in range(len(bins)-1):
  x.append(str(bins[i]) + "-" + str(bins[i+1]))
for i in range(len(bins)-1):
  y.append(counts[0][i]/counts[1][i] * 100)

plt.plot(x, y, "bo--")
plt.xlabel('Shot Clock (s)')
plt.ylabel('Shot Percentage')
plt.title('Shot Percentage Over a 24 Second Period')
plt.show()
```
